## Task 3.7: Try concatenating x and v (different shapes)

```python
print("Trying to concatenate x and v (different shapes):")
print(f"x shape: {x.shape}")
print(f"v shape: {v.shape}")

try:
    result = np.concatenate([x, v])
    print(result)
except ValueError as e:
    print(f"\n ERROR: {e}")
    print("\n Explanation:")
    print("x has shape (2,2) and v has shape (2,).")
    print("They have different number of dimensions, so they cannot be")
    print("concatenated along axis=0.")
    print("\n Solution: Reshape v to (1,2) or use vstack instead.")
    result = np.vstack([x, v])
    print("\nUsing np.vstack:")
    print(result)
```

# PROBLEM 4: MATRIX OPERATIONS

## Define matrices for Problem 4

```python
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
```

## Task 4.1: Prove A ¬∑ A^(-1) = I

```python
A_inverse = np.linalg.inv(A)
print("A^(-1) (inverse of A):")
print(A_inverse)

result = np.dot(A, A_inverse)
print("\nA ¬∑ A^(-1):")
print(result)

print("\nIdentity Matrix (rounded):")
print(np.round(result))
print("\n A ¬∑ A^(-1) = I (Identity Matrix)")
```

```python
print("System of equations:")
print("3x + y - z = 9")
print("2x - 3y + z = -1")
print("x - y + 2z = -3")

coeff_matrix = np.array([[3, 1, -1],
                         [2, -3, 1],
                         [1, -1, 2]])

constants = np.array([9, -1, -3])

solution = np.linalg.solve(coeff_matrix, constants)

print("\nSolution:")
print(f"x = {solution[0]:.4f}")
print(f"y = {solution[1]:.4f}")
print(f"z = {solution[2]:.4f}")

print("\n Verification:")
eq1 = 3*solution[0] + solution[1] - solution[2]
eq2 = 2*solution[0] - 3*solution[1] + solution[2]
eq3 = solution[0] - solution[1] + 2*solution[2]
print(f"3x + y - z = {eq1:.4f} (should be 9)")
print(f"2x - 3y + z = {eq2:.4f} (should be -1)")
print(f"x - y + 2z = {eq3:.4f} (should be -3)")
```

## Experiment 3: Dot Product

```python
size_matrix = 1000

print(f"Comparing Python lists vs NumPy for matrix multiplication")
print(f"Matrix size: {size_matrix}√ó{size_matrix}\n")

python_matrix_1 = [[i+j for j in range(size_matrix)] for i in range(size_matrix)]
python_matrix_2 = [[i+j for j in range(size_matrix)] for i in range(size_matrix)]

print("‚è≥ Computing Python lists (this may take a while)...")
start = time.time()
python_matmul = [[sum(python_matrix_1[i][k] * python_matrix_2[k][j]
                      for k in range(size_matrix))
                  for j in range(size_matrix)]
                 for i in range(size_matrix)]
python_time = time.time() - start

print(f"Python lists: {python_time:.6f} seconds")

numpy_matrix_1 = np.random.rand(size_matrix, size_matrix)
numpy_matrix_2 = np.random.rand(size_matrix, size_matrix)

print("‚è≥ Computing NumPy arrays...")
start = time.time()
numpy_matmul = np.dot(numpy_matrix_1, numpy_matrix_2)
numpy_time = time.time() - start

print(f"NumPy arrays: {numpy_time:.6f} seconds")
print(f"\n NumPy is {python_time / numpy_time:.2f}x FASTER!")
print("\n This is why NumPy is so important for data science!")
```
