```python
import matplotlib.pyplot as plt
```

```python
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
low= []
moderate= []
high= []
for study,_,_ in time_data:
    if study < 3.0:
        low.append((study,_,_))
    elif 3.0 <= study <= 5.0:
        moderate.append((study,_,_))
    else:
        high.append((study,_,_))
print("Low study time entries:", low)
print("Moderate study time entries:", moderate)
print("High study time entries:", high)
```

```python
print(f"{len(low)} Low study time days ")
print(f"{len(moderate)} Moderate study time days ")
print(f"{len(high)} High study time days ")
```

```python
study_minutes = []
for study, _, _ in time_data:
    study_minutes.append(study * 60)
print( study_minutes)
```

```python
import matplotlib.pyplot as plt

study_hours = []
entertainment_hours = []
sleep_hours = []
for study, entertainment, sleep in time_data:
    study_hours.append(study)
    entertainment_hours.append(entertainment)
    sleep_hours.append(sleep)

count_study = len(study_hours)
count_entertainment = len(entertainment_hours)
count_sleep = len(sleep_hours)

sum_study = sum(study_hours)
sum_entertainment = sum(entertainment_hours)
sum_sleep = sum(sleep_hours)

avg_study = sum_study / count_study
avg_entertainment = sum_entertainment / count_entertainment
avg_sleep = sum_sleep / count_sleep
print(f"Average study time: {avg_study} hours")
print(f"Average entertainment time: {avg_entertainment} hours")
print(f"Average sleep time: {avg_sleep} hours")
```

```python
plt.scatter(study_hours, sleep_hours, color='blue', s=100, alpha=0.6)
plt.title('Study Hours vs Sleep Hours', color='darkblue', fontsize=16, fontweight='bold')
plt.xlabel('Study Hours', color='darkgreen', fontsize=12)
plt.ylabel('Sleep Hours', color='darkred', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def sum_nested_list(nested_list):

    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

print(sum_nested_list([1, [2, [3, 4], 5], 6, [7, 8]]))
print(sum_nested_list([1, [2, 3], [4, [5]]]))
```

```python
def generate_permutations(s):

    if len(s) <= 1:
        return [s]

    result = []


    for i in range(len(s)):
        current_char = s[i]
        remaining_chars = s[:i] + s[i+1:]

        for perm in generate_permutations(remaining_chars):
            result.append(current_char + perm)

    return list(set(result))
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    total_size = 0
    for name, value in directory.items():
        if isinstance(value, int):
            total_size += value
        else:
            total_size += calculate_directory_size(value)

    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

print(calculate_directory_size(directory_structure))
```

```python
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"Coin: {coins}, Amount: {amount}")
print(f"Minimum amount of   coins needed: {result}")
```

```python
def longest_common_subsequence(s1, s2):
    n = len(s1)
    m = len(s2)
    dp = [[0]*(m+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for j in range(1, m+1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])

    return dp[n][m]

print(longest_common_subsequence("abcde", "ace"))
```

```python
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0]*(capacity+1) for _ in range(n+1)]

    for i in range(1, n+1):
        w = weights[i-1]
        v = values[i-1]
        for c in range(1, capacity+1):
            if w <= c:
                dp[i][c] = max(v + dp[i-1][c-w], dp[i-1][c])
            else:
                dp[i][c] = dp[i-1][c]

    return dp[n][capacity]

print(knapsack([1,3,4,5], [1,4,5,7], 7))
```
